```python
import matplotlib.pyplot as plt
import numpy as np
```

line plot- used for the plotting 2 points/values , to represent the relation b/w them..
* Key Attributes
1.  color – Changes the line color.
2.  linestyle – '-', '--', ':', 'None'
3.  linewidth – Thickness of the line.
4.  marker – 'o', '*', 's', '^', etc.
5.  markersize – Size of the marker.
6.  markerfacecolor – Fill color of the marker.


```python
#line plot- used for the plotting 2 points/values , to represent the relation b/w them..
hours=[1,2,3,4,5,6,7,8]
visitors=[50,40,20,56,23,34,54,67]
plt.plot(hours,visitors,color='blue',marker='o',linestyle=':',label='Visitors',markerfacecolor='yellow')
plt.legend() #lengend takes parmeter label..
plt.title("Website Visitors Per Hour")
plt.xlabel("Hours")
plt.ylabel("Visitors")
plt.grid(True)
plt.show()
```

Bar Chart is used for the represent the data in the bars/blocks
*  Key Attributes
1.   color – Changes bar color.
2.   edgecolor – Sets border color.
3.   linewidth – Sets bar border thickness.
4.   alpha – Controls transparency (0 to 1).


```python
products=['Laptop','Mobile','Tablet','Headphones','Smartwatch']
sales=[250,400,150,300,180]
plt.bar(products,sales,color='orange',edgecolor='black',alpha=1)
plt.title('Product Sales Comparision')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

Histogram

Used for distribution analysis.
*   Key Attributes
1.  bins – Defines the number of bins.
2.  color – Sets bar color.
3.  edgecolor – Defines bar edges.
4.  alpha – Adjusts transparency.


```python
marks=[2, 9, 13, 17, 21, 28, 32, 37, 43, 46,
 54, 58, 60, 67, 70, 73, 82, 84, 88, 91,
 93, 96, 26, 35, 40, 56, 64, 69, 80, 98]


plt.hist(marks,bins=10,color='green',edgecolor='black',alpha=1)
plt.title('StudentMarks Distribution')
plt.xlabel('Marks Range')
plt.ylabel('number of student')
plt.show()
```

Scatter Plot

Used for correlation between two variables.
Key Attributes:
1. s – Changes point size.
2. c – Assigns color using values.
3. cmap – Uses a colormap ('viridis', 'plasma', 'coolwarm').
4. alpha – Adjusts transparency.
5. edgecolors – Defines the border of the points.







```python
ad_spend= [2,4,5,7,8,10,11,12]
sales=[20,25,28,35,40,48,50,52]
sizes=[200,250,280,350,400,480,500,520]
colors=sales
plt.scatter(ad_spend,sales,s=sizes,c=colors,cmap='viridis',alpha=0.8,edgecolors='black',linewidths=1.2,label='sales')
plt.title('Advertising Spend vs Sales',fontsize=14,fontweight='bold')
plt.xlabel('Ad Spend')
plt.ylabel('Sales')
plt.grid(True,linestyle='--',alpha=0.5)
plt.colorbar(label='Sales Value')
plt.legend()
plt.show()
```

Pie Chart
Key Attributes
1. autopct – Displays percentages.
2. startangle – Rotates the chart.
3. explode – Separates a slice for emphasis.

```python
brands=['Apple','Samsung','Oppo','Realme','Others']
market_share=[30,25,20,10,15]
plt.pie(market_share,labels=brands,autopct='%1.2f%%',startangle=90,explode=(0.1,0,0,0,0),shadow=True)
plt.title('Mobile Market Share')
plt.show()
```

Box Plot
 Used for statistical distribution.
Key Attributes
1.   patch_artist – Fills the box with color.
2.   boxprops – Modifies box appearance.



```python
scores=[
    [78,85,90,88,76,95,9],
    [65,70,68,72,75,80,85,78],
    [90,92,94,96,88,91,89,93],
    [62,60,61,90,65,63,67,61],
    [70,85,8,50,92,91,100,89]
]
plt.boxplot(scores,patch_artist=True,notch=False,vert=True,boxprops=dict(facecolor='blue',color='black'),labels=['A','B','C','D','E'])
plt.title('Exam Scores Distribution Across Classes')
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.grid(True,linestyle=':',alpha=0.7)
plt.show()
```
